# Standardize categorical data

```python
df = df.drop_duplicates()
```

```python
df['BP_History'] = df['BP_History'].map({'Yes':1, 'No':0})

df['Medication'] = df['Medication'].map({'Yes':1, 'No':0})

df['Family_History'] = df['Family_History'].map({'Yes':1, 'No':0})

df['Smoking_Status'] = df['Smoking_Status'].map({'Yes':1, 'No':0})

df['Has_Hypertension'] = df['Has_Hypertension'].map({'Yes':1, 'No':0})
```

#Exercise level encoding

```python
df['Exercise_Level'] = df['Exercise_Level'].map({'Low':0, 'Moderate':1, 'High':2})
```

# Corelation Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```
